```python
from __future__ import print_function
from hyperopt import Trials, STATUS_OK, tpe
from hyperas import optim
from hyperas.distributions import choice, uniform

from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD

from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import time
```

```python
seed = 7
np.random.seed(seed)
```

```python
df = pd.read_csv('car.csv')
print(df.shape)
type(df)  # pandas.core.frame.DataFrame.
df.head()
```

```python
def data():
    df = pd.read_csv('car.csv')
    X = df.values[:, :-1]
    Y = pd.get_dummies(df['Class_Val']).values
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=0)
    return X_train, X_test, y_train, y_test
```

```python
def model(X_train, X_test, y_train, y_test):
    model = Sequential()
    model.add(
        Dense(
            {{choice([8, 16, 32, 64, 128])}},
            input_shape=(6, ),
            kernel_initializer={{
                choice([
                    'uniform', 'lecun_uniform', 'normal', 'zero',
                    'glorot_normal', 'glorot_uniform', 'he_normal',
                    'he_uniform'
                ])
            }}))
    model.add(
        Activation({{
            choice([
                'softplus', 'softsign', 'relu', 'tanh', 'sigmoid',
                'hard_sigmoid', 'linear'
            ])
        }}))
    model.add(Dropout({{uniform(0, 0.5)}}))

    model.add(
        Dense(
            {{choice([8, 16, 32, 64, 128])}},
            kernel_initializer={{
                choice([
                    'uniform', 'lecun_uniform', 'normal', 'zero',
                    'glorot_normal', 'glorot_uniform', 'he_normal',
                    'he_uniform'
                ])
            }}))
    model.add(
        Activation({{
            choice([
                'softplus', 'softsign', 'relu', 'tanh', 'sigmoid',
                'hard_sigmoid', 'linear'
            ])
        }}))
    model.add(Dropout({{uniform(0, 0.5)}}))

    model.add(
        Dense(
            4,
            kernel_initializer={{
                choice([
                    'uniform', 'lecun_uniform', 'normal', 'zero',
                    'glorot_normal', 'glorot_uniform', 'he_normal',
                    'he_uniform'
                ])
            }},
            activation='softmax'))

    model.compile(
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        optimizer={{
            choice([
                'SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax',
                'Nadam'
            ])
        }})

    model.fit(
        X_train,
        y_train,
        batch_size={{choice([16, 32, 64, 128, 256])}},
        epochs={{choice([10,30,50,100])}},
        verbose=2,
        validation_data=(X_test, y_test))
    
    score, acc = model.evaluate(X_test, y_test, verbose=0)
    
    print('Test accuracy:', acc)
    return {'loss': -acc, 'status': STATUS_OK, 'model': model}
```

```python
X_train, X_test,y_train,y_test = data()
```

```python
start = time.time()
best_run, best_model = optim.minimize(model=model,
                                      data=data,
                                      algo=tpe.suggest,
                                      max_evals=300,
                                     trials=Trials(),
                                      notebook_name='19_Hyperparamete_Opti_TPE_Classification_Car')
print('@ Total Time Spent: %.2f seconds' % (time.time()-start))
```

```python
print("Evalutation of best performing model:")
print(best_model.evaluate(X_test, y_test))
print("Best performing model chosen hyper-parameters:")
print(best_run)
```
