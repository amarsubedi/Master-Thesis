## 1. Plot Sigmoid Funciton and it's Derivative

```
#import section
from matplotlib import pylab
import pylab as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-white')
```

```
#sigmoid = lambda x: 1 / (1 + np.exp(-x))
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

# Derivative of sigmoid
def der_sigmoid(x):
  return sigmoid(x)*(1-sigmoid(x))

# generate an Array with value ???
# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
# 
x = plt.linspace(-10,10,100)

fig, ax = plt.subplots(figsize=(10,5))
# prepare the plot, associate the color r(ed) or b(lue) and the label 
ax.plot(x, sigmoid(x), 'b',linewidth=3,label='Sigmoid Function')
ax.plot(x, der_sigmoid(x),linestyle=":",color='red',linewidth=5,label='Derivative Sigmoid')
# Title & Subtitle
plt.title("Sigmoid Function and its Derivative",fontsize=14, fontweight='bold')

# place the legen boc in bottom right of the graph
ax.legend(loc='best',frameon=True,edgecolor='black',borderpad=1)

# write the Sigmoid formula
#plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)

#resize the X and Y axes
#plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 

# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
ax.set_facecolor('w')
ax.grid(True, color='black',linestyle='-',linewidth=1,alpha=0.3)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

## 2. Plot Tanh Funciton and it's Derivative

```
def tanh(x):
    return np.tanh(x)

# Derivative of sigmoid
def der_tanh(x):
  return 1- tanh(x)*tanh(x)

# generate an Array with value ???
# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
# 
x = plt.linspace(-10,10,100)

fig, ax = plt.subplots(figsize=(10,5))
# prepare the plot, associate the color r(ed) or b(lue) and the label 
ax.plot(x, tanh(x), 'b',linewidth=3,label='Tanh Function')
ax.plot(x, der_tanh(x),linestyle=":",color='red',linewidth=5,label='Derivative Tanh')
# Title & Subtitle
plt.title("Tanh Function and its Derivative",fontsize=14, fontweight='bold')

# place the legen boc in bottom right of the graph
ax.legend(loc='best',frameon=True,edgecolor='black',borderpad=1)


# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
ax.set_facecolor('w')
ax.grid(True, color='black',linestyle='-',linewidth=1,alpha=0.3)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

## 3. Plot ReLU Funciton and it's Derivative

```
# define ReLU
def Relu(x):
    """ReLU returns 1 if x>0, else 0."""
    return np.maximum(0,x)
  
# define it's derivative
def der_relu(x):
  return np.greater(x, 0).astype(int)
  

x = plt.linspace(-5,5,100)

fig, ax = plt.subplots(figsize=(10,5))
# prepare the plot, associate the color r(ed) or b(lue) and the label 
ax.plot(x, Relu(x), 'b',linewidth=3,label='ReLU Function')
ax.plot(x, der_relu(x),linestyle=":",color='red',linewidth=5,label='Derivative ReLU')
ax.plot(0, Relu(0), 'og',markersize=12,label="Undefined but Derivative=0")
# Title & Subtitle
plt.title("ReLU Function and its Derivative",fontsize=14, fontweight='bold')

# place the legen boc in bottom right of the graph
ax.legend(loc='best',frameon=True,edgecolor='black',borderpad=1)


# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
ax.set_facecolor('w')
ax.grid(True, color='black',linestyle='-',linewidth=1,alpha=0.3)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

## 4. Plot LeakyReLU Funciton and it's Derivative

```
# define LeakyReLU
def LeakyRelu(x,alpha=0.1):
    """ReLU returns 1 if x>0, else 0."""
    return np.maximum(alpha*x,x)
  
# define it's derivative
def der_leakyrelu(x, alpha=0.1):
  dx = np.ones_like(x)
  dx[x < 0] = alpha
  return dx
  

x = plt.linspace(-5,5,100)

fig, ax = plt.subplots(figsize=(10,5))
# prepare the plot, associate the color r(ed) or b(lue) and the label 
ax.plot(x, LeakyRelu(x,0.1), 'b',linewidth=3,label='LeakyReLU Function')
ax.plot(x, der_leakyrelu(x,0.1),linestyle=":",color='red',linewidth=5,label='Derivative LeakyReLU')
ax.plot(0, LeakyRelu(0), 'og',markersize=12,label="Undefined but Derivative=0")
# Title & Subtitle
plt.title("LeakyReLU Function and its Derivative",fontsize=14, fontweight='bold')

# place the legen boc in bottom right of the graph
ax.legend(loc='best',frameon=True,edgecolor='black',borderpad=1)

# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
ax.set_facecolor('w')
ax.grid(True, color='black',linestyle='-',linewidth=1,alpha=0.3)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```

## 5. Plot Identity Funciton and it's Derivative

```
# define LeakyReLU
def identity(x):
    return x
def der_identity(x):
  return np.ones(x.shape)

x = plt.linspace(-5,5,100)

fig, ax = plt.subplots(figsize=(10,5))
# prepare the plot, associate the color r(ed) or b(lue) and the label 
ax.plot(x, identity(x), 'b',linewidth=3,label='Identity Function')
ax.plot(x, der_identity(x),linestyle=":",color='red',linewidth=5,label='Derivative Identity')

# Title & Subtitle
plt.title("Identity Function and its Derivative",fontsize=14, fontweight='bold')

# place the legen boc in bottom right of the graph
ax.legend(loc='best',frameon=True,edgecolor='black',borderpad=1)

# plt.plot(x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# create the graph
ax.set_facecolor('w')
ax.grid(True, color='black',linestyle='-',linewidth=1,alpha=0.3)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.show()
```
