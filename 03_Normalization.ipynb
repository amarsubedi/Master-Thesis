```
import pandas as pd
import numpy as np
import io
```

```
df = pd.io.parsers.read_csv(
    'https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
     header=None,
     usecols=[0,1,2]
    )

df.columns=['Class label', 'Alcohol', 'Malic acid']

df.head()
```

## Standardization and Min-Max Scaling

```
from sklearn import preprocessing
```

```
std_scale = preprocessing.StandardScaler().fit(df[['Alcohol', 'Malic acid']])
df_std = std_scale.transform(df[['Alcohol','Malic acid']])
```

```
minmax_scale = preprocessing.MinMaxScaler().fit(df[['Alcohol','Malic acid']])
df_minmax = minmax_scale.transform(df[['Alcohol','Malic acid']])
```

```
%matplotlib inline
```

```
from matplotlib import pyplot as plt
plt.style.use('seaborn-white')

fig,ax = plt.subplots()
ax.scatter(df['Alcohol'],df['Malic acid'],color='green',label='Original data', alpha=0.8)
ax.scatter(df_std[:,0], df_std[:,1], color='blue',label='Standardized', alpha=0.8)
ax.scatter(df_minmax[:,0],df_minmax[:,1], color='red',label='Min-Max scaled', alpha=0.8)


plt.title('Comparison of Normalization ',fontsize=20,fontname="sans-serif")
plt.xlabel('x',fontsize=18,fontname="sans-serif")
plt.ylabel('y',fontsize=18,fontname="sans-serif")

ax.legend(loc='upper left',frameon=True,edgecolor='black')
ax.grid(True, color='black',linestyle='dotted',linewidth=1,alpha=0.3)

plt.show()
```
